```python
import pandas as pd
from matplotlib import pyplot as plt

```

```python
#generating the table column names
Feature_columnnnames=[]
for i in range(48):
    Feature_columnnnames.append("Feature"+str(i+1))
Feature_columnnnames.append("label")
```

```python
dataset = pd.read_csv('C:\\Users\\KARTIK\\Desktop\\Sensorless_drive_diagnosis.txt', sep=" ", header=None,names=Feature_columnnnames)

# randomizing data and collecting only 10000 data
dataset = dataset.sample(frac = 1)
dataset = dataset.head(10000)
print(dataset)
```

```python
x = dataset[{'Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5', 'Feature6', 'Feature7', 'Feature8', 'Feature9', 'Feature10', 'Feature11', 'Feature12', 'Feature13', 'Feature14', 'Feature15', 'Feature16', 'Feature17', 'Feature18', 'Feature19', 'Feature20', 'Feature21', 'Feature22', 'Feature23', 'Feature24', 'Feature25', 'Feature26', 'Feature27', 'Feature28', 'Feature29', 'Feature30', 'Feature31', 'Feature32', 'Feature33', 'Feature34', 'Feature35', 'Feature36', 'Feature37', 'Feature38', 'Feature39', 'Feature40', 'Feature41', 'Feature42', 'Feature43', 'Feature44', 'Feature45', 'Feature46', 'Feature47', 'Feature48'}]
x = x.values.tolist()
y = list(dataset['label'])
print(y)
```

```python
# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.45, random_state=1)
print(y_test)
```

```python
# training the model on training set
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
  
# making predictions on the testing set
y_pred = gnb.predict(x_test)
print(y_pred)
```

```python
# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```

```python
# Constructing the confusion matrix.
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
import numpy as np
from sklearn .metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# converted y_pred to list to keep comman between numbers
y_pred = list(y_pred)

# ROC Curve is made when output is either 1 or 0 i.e. true or false
# But in our case, output can be from 1 to 11 so we need 11 ROC Curves for each output value from 1 to 11
# For each output i, we will make its corresponding arrays with only 1 and 0 values, 1 if both actual and predicted values are i otherwise 0 

# Compute ROC curve and ROC area for each class
n_classes = 11
for i in range(1, n_classes + 1):
    y_test1 = [0]
    y_pred1 = [0]
    for j in range(len(y_test)):
        if y_test[j] == i and y_pred[j] == i:
            y_test1.append(1)
            y_pred1.append(1)
        if y_test[j] == i and y_pred[j] != i:
            y_test1.append(1)
            y_pred1.append(0)
        if y_test[j] != i and y_pred[j] == i:
            y_test1.append(0)
            y_pred1.append(1)    
            
    fpr, tpr, threshold = metrics.roc_curve(y_test1, y_pred1)
    roc_auc = metrics.auc(fpr, tpr)

    plt.title('ROC Curve for Class ' + str(i))
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()

```
